```r
library(Seurat)
library(dplyr)
library(ggplot2)
```

```r
total <- readRDS("/home/datastore2/karlso/datastore/scRNAseq_GSE155121_human_embryo_CS10to14/20230812_withvelo/Processed_data/scRNA/Step1f_cleaned_annotated.rds")
```

```r
options(repr.plot.width=10, repr.plot.height=10)
DimPlot(total, group.by = 'celltype', reduction = "draw_graph_fa", label = TRUE) & NoAxes() & NoLegend()

Idents(total) <- "celltype"
levels(total) <- c("01_Epiblast",
                  '02_Non_neuroectoderm',
                  '03_Epidermal',
                  '04_Placode',
                  '05_Neural_crest',
                  '06_Neuroectoderm',
                  '07_Telencephalon',
                  '08_Diencephalon',
                  '09_Mesencephalon',
                  '10_Rhombomere',
                  '11_Roof_plate')
```

```r
options(repr.plot.width=4, repr.plot.height=4, repr.plot.res =200)
DimPlot(total, group.by = "celltype", reduction = "draw_graph_fa", label = T, order = T) & NoAxes() & NoLegend()
DimPlot(total, group.by = "celltype", reduction = "draw_graph_fa", label = F, order = T) & NoAxes() & NoLegend()
FeaturePlot(total, features = "ISL1", reduction = "draw_graph_fa", order = T, cols = c("lightgrey", "red")) & NoAxes() & NoLegend()
FeaturePlot(total, features = "PRDM1", reduction = "draw_graph_fa", order = T, cols = c("lightgrey", "red")) & NoAxes() & NoLegend()
```
