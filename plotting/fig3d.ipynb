```r
suppressMessages(library(tidyverse))
suppressMessages(library(data.table))
library(UpSetR)
```

```r
data <- fread('data/candidate-gene-annotation.txt', header = T)
```

```r
head(data)
```

```r
data <- data %>% column_to_rownames(var='gene')
```

```r
data$unknown_candidate <- ifelse(data$human_database == 0 & data$mouse_database == 0, 1, 0)
```

```r
options(repr.plot.width = 10, repr.plot.height = 5, repr.plot.res = 500)
p <- upset(data, show.numbers = 'yes', sets = c('unknown_candidate', 'placode_expressed', 'ncc_expressed', 'mouse_database', 'human_database'),
           keep.order = T, order.by = "freq")
p
```

```r
svg(file="figures/upset_candidate-annotation_v1-2_2025q4.svg", width = 10, height = 5) # or other device
p
dev.off()
```
