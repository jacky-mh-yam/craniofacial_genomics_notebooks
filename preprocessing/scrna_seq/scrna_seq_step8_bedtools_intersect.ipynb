# Intersect DEGs with GWAS Summary

```python
%%bash
conda run -n bedtools bash -c '
celltypes=(
    epiblast
    non_neuroectoderm
    epidermal
    placode
    neural_crest
    neuroectoderm
    telencephalon
    diencephalon
    mesencephalon
    rhombomere
    roof_plate
)

for celltype in "${celltypes[@]}"; do
    bedtools intersect -wao \
        -a ../snp_array/outputs/GWAS_Summary.bed \
        -b data/${celltype}_400kb.bed \
        > outputs/gwas_summary_${celltype}_intersect.bed
done'
```
