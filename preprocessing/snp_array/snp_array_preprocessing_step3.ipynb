# Preprocessing SNP array data - step3

```r
suppressMessages(library(data.table))
suppressMessages(library(tidyverse))
```

## Generate GWAS summary file

```r
# Read and filter logistic regression results
gwas_raw <- fread("data/GWAS_LogisticRegression.assoc.logistic")
gwas_add <- gwas_raw[gwas_raw$TEST == "ADD", ] %>% select(CHR, SNP, BP, P)
```

```r
# Inspect filtered results
summary(gwas_add$P)
nrow(gwas_add)
head(gwas_add)
```

```r
# Convert to BED format: CHR, Start, End, SNP, P
gwas_bed <- gwas_add[, c("CHR", "BP", "BP", "SNP", "P")]
colnames(gwas_bed) <- c("CHR", "Start", "End", "SNP", "P")
```

```r
# Export
write.table(gwas_bed, file = "outputs/GWAS_Summary.bed", sep = "\t", row.names = FALSE, col.names = FALSE, quote = FALSE)
```
